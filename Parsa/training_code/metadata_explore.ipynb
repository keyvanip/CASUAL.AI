```python
import json
from collections import Counter

# Load metadata
with open("../metadata.json") as f:
    metadata = json.load(f)

# Initialize a color counter
color_counter = Counter()

# Loop through all items in each category
for category, items in metadata.items():
    for item in items:
        color_field = item.get("color", [])
        
        # Ensure it's a list, even if it's a string
        if isinstance(color_field, str):
            colors = [color_field.lower()]
        elif isinstance(color_field, list):
            colors = [c.lower() for c in color_field if isinstance(c, str)]
        else:
            colors = []

        # Count all valid colors
        color_counter.update(colors)

# Display results
print("Color Distribution in metadata.json file:\n")
for color, count in color_counter.most_common():
    print(f"{color}: {count}")
```

```python
import json
from collections import Counter

# Load metadata
with open("../metadata.json") as f:
    metadata = json.load(f)

# Initialize a dictionary to hold color counters for each category
category_color_counts = {
    "pants": Counter(),
    "shorts": Counter(),
    "shirt": Counter(),
    "t-shirt": Counter()
}

# Loop through each category and count colors
for category in category_color_counts.keys():
    items = metadata.get(category, [])
    for item in items:
        color_field = item.get("color", [])

        if isinstance(color_field, str):
            colors = [color_field.lower()]
        elif isinstance(color_field, list):
            colors = [c.lower() for c in color_field if isinstance(c, str)]
        else:
            colors = []

        category_color_counts[category].update(colors)

# Display results
for category, counter in category_color_counts.items():
    print(f"\nðŸŽ¨ Color Distribution for '{category}':")
    for color, count in counter.most_common():
        print(f"  {color}: {count}")
```
